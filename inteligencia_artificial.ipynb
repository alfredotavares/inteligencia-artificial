```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('listings.csv')
print(len(data))

data.fillna(method ='ffill', inplace = True)
data.head()
```

```python
print(len(data['neighbourhood'].unique()))
print(len(data['host_name'].unique()))
print(len(data['room_type'].unique()))
```

```python
features = ['room_type', 'minimum_nights', 'reviews_per_month', 'price']
```

```python
df = data[features]
df
```

```python
LE_room = LabelEncoder()
df['room_type'] = LE_room.fit_transform(df['room_type'])
```

```python
df
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
# Dropping any rows with Nan values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
  
# Splitting the data into training and testing data
regr = LinearRegression()
  
regr.fit(X_train, y_train)
print(regr.score(X_test, y_test))
```
